```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import csv
import numpy
```

```python
dados_cnpj = pd.DataFrame(columns=['CNPJ','CNAES1']) #cria um frame de dados
df = pd.read_csv(r"C:\Users\All In One\Desktop\coletar cnaes\listacnpj.csv") #lê um documento csv com vários cnpjs
unique_arr = df["cnpj"].unique() #seleciona a coluna cnpj e transforma em array
print(unique_arr.size)
dados_cnpj.head(20)


    
    








```

```python
for i in unique_arr:
    
    cnpj = i #para cada elemento na array "unique_arr"
    endereço = "https://cnpj.biz/" # variável que guarda o endereço do site
    url = (endereço + str(cnpj)) # variável que junta o prefixo(site) com o cnpj para consulta
    payload = {'api_key': 'f3420a407a52a05a2180a5377bfdee15', 'url': url}
    data = requests.get('http://api.scraperapi.com', params=payload).text
    
    #data = requests.get(url).text # pede para o site as informações contidas no endereço

    soap = bs(data,'html5lib') #passa o conteúdo do site para ser lido pelo BS
    cnaes = str(soap.find_all('u')) #encontra todas as tags <u> --o site coloca os CNAES em tags <u>
    #cnaes_split = cnaes.lower().split(", ",1)
    #vcnaes_arr = numpy.array(cnaes_split)
    
         
    dados_cnpj = dados_cnpj.append({'CNPJ':cnpj,'CNAES1':cnaes},ignore_index=True) # Passa para o quadro de dados o cnpj e a informação coletada

    
```

```python
dados_cnpj.head(20)
```

```python
dados_limpos = dados_cnpj.copy()

dados_limpos['CNAES2'] = dados_limpos['CNAES1'].str.split(' - ', n=2, expand=True)[1]
dados_limpos['CNAES1'] = dados_limpos['CNAES1'].str.strip()
dados_limpos['CNAES1'] = dados_limpos['CNAES1'].str.split(' - ', expand=True)[0].str.split(r'\<u>', expand=True)[1]
dados_limpos['CNAES2'] = dados_limpos['CNAES2'].str.split('</u>', expand=True)[0]
dados_limpos['CNAES2'][1]


```

```python

dados_limpos.to_csv(r'C:\Users\All In One\Desktop\coletar cnaes\scrapresult.csv',encoding='utf-8-sig') #Grava o frame de dados em um arquivo csv
    
```
